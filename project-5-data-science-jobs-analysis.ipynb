**EXPLORATORY DATA ANALYSIS**

```python
jd.columns
```

```python
top10_companies = jd['Company'].value_counts()[:10]
fig = px.bar(y=top10_companies.values, 
             x=top10_companies.index, 
             color = top10_companies.index,
             color_discrete_sequence=px.colors.sequential.turbid,
             text=top10_companies.values,
             title= 'Top 10 Companies',
             template= 'plotly_dark')
fig.update_layout(xaxis_title="Companies",yaxis_title="count",font = dict(size=17,family="Franklin Gothic"))
fig.show()
```

**Here comes the result of top 10 companies.**

```python
top10_roles = jd['JobTitle'].value_counts()[:10]
fig = px.bar(y=top10_roles.values, 
             x=top10_roles.index, 
             color = top10_roles.index,
             color_discrete_sequence=px.colors.sequential.OrRd,
             text=top10_roles.values,
             title= 'Top 10 JobTitle',
             template= 'plotly_dark')
fig.update_layout(
    xaxis_title="JobTitle",
    yaxis_title="count",
    font = dict(size=17,family="Franklin Gothic"))
fig.show()
```

**Here is the plot of top 10 JobTitle among which Data Scientist secures the topmost level.**

```python
my_explode=(0,0,0.1,0.1,0.1) 
labels= [x for x in jd["Company"].value_counts().sort_values(ascending=False)[0:5].index] 
patches,ax, text= plt.pie(jd["Company"].value_counts().sort_values(ascending=False)[0:5],autopct='%1.1f%%',explode=my_explode,shadow=True,startangle=305)
plt.title("Top 5 Companies with Most Number of Data Science Related Jobs", size=15, color='black')
plt.legend(patches, labels, loc=4, bbox_to_anchor = (1.3,0.25))
plt.axis('equal')
plt.show()
```
